### **Univariate Analysis**

```python
bins = [0, 100000, 300000, np.inf]
labels = ["Low", "Medium", "High"]

df["Income Category"] = pd.cut(df["Estimated Income"], bins=bins, labels=labels)
```

```python
plt.pie(
    df["Income Category"].value_counts(),
    labels=df["Income Category"].value_counts().index,
    autopct="%1.1f%%", 
    startangle=90
)
plt.title("Distribution of Income Categories")
plt.axis("equal")  
plt.show()
```

## **Bivariate Analysis: Hue-> Gender** 

```python
# Examine the distribbution of unique categories in categorical columns
categorical_columns = df[["BRId","GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Category"]]
for i, predictor in enumerate(categorical_columns.columns):
    plt.figure(i)
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x=predictor, hue="GenderId")
    plt.title(f"Distribution of {predictor}")
    plt.xticks(rotation=45)
    plt.show()
     
```

```python
numerical_columns = df[["Estimated Income", "Credit Card Balance", "Checking Accounts", "Superannuation Savings", "Bank Loans", "Bank Deposits", "Saving Accounts", "Foreign Currency Account", "Business Lending"]]
```

```python
for i, predictor in enumerate(numerical_columns.columns):
    plt.figure(i)
    plt.figure(figsize=(8, 5))
    sns.histplot(data=df, x=predictor, kde=True, bins=30)
    plt.title(f"Distribution of {predictor}")
    plt.show()
```

```python
correlation_matrix = numerical_columns.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Numerical Features")
plt.show()
```

- If a person is holding large amount of money in one account then it is likely to hold a substantial amount of money in other accounts too.

- Credit usage pattern is very much independent of income levels and bank loans.
